# 重點:
取出詳細內容的文字
問題:不確定其他頁數要的資料的div位置也相同

```python
# 找出內如demo2,因每頁項目不同
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
from selenium.webdriver import ActionChains
from bs4 import BeautifulSoup
import csv
s = Service("./chromedriver.exe")
driver = webdriver.Chrome(service=s)
driver.get("https://www.yes123.com.tw/wk_index/job.asp?p_id=20110211160133_28771443&job_id=20140124095456_13304731")
time.sleep(2)
# titles=driver.find_elements_by_class_name("Job_opening_item")#舊寫法
job_detail = driver.page_source
driver.close()
print(job_detail)
```

```python
soup = BeautifulSoup(str(job_detail), 'html.parser')
# print(soup)
# print(soup.prettify())
# print(soup)
```

```python
#工作技能 loop 取出文字
job_key_div = soup.find_all("div",{"class":"job_explain"})[4].find_all("span",{"class":"right_main"})
for i in job_key_div:
    print(i.get_text())
```

```python
#薪資
job_salary_div = soup.find("div",{"class":"job_explain mt"}).find_all("span",{"class":"right_main"})[1]
# for i in job_content_div:
#     print(i.get_text())
job_salary_div.get_text()
```

```python
#工作內容
job_content_div = soup.find("div",{"class":"job_explain mt"}).find_all("span",{"class":"right_main"})[0]
# for i in job_content_div:
#     print(i.get_text())
job_content_div.get_text()
```

```python
#地址
job_addr_div = soup.find("div",{"class":"job_explain mt"}).find_all("span",{"class":"right_main"})[5].find("a")
# for i in job_content_div:
#     print(i.get_text())
job_addr_div.get_text()
```

```python
#學歷
job_sch_div = soup.find_all("div",{"class":"job_explain"})[2].find_all("span",{"class":"right_main"})[0]
# for i in job_content_div:
#     print(i.get_text())
job_sch_div.get_text()
```

```python
#科系要求
job_dep_div = soup.find_all("div",{"class":"job_explain"})[2].find_all("span",{"class":"right_main"})[1]
# for i in job_content_div:
#     print(i.get_text())
job_dep_div.get_text()
```

```python
# 擅長工具
job_can_div = soup.find_all("div",{"class":"job_explain"})[4].find_all("span",{"class":"right_main"})
for i in job_can_div:
    print(i.get_text())
# job_content_div.get_text()
```

```python
# 其他條件
job_other_div = soup.find_all("div",{"class":"job_explain"})[5].find_all("span",{"class":"exception"})
for i in job_other_div:
    print(i.get_text())
# job_content_div.get_text()
```

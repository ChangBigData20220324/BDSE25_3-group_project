### 爬取職位名稱
 - 方法
用迴圈爬取相同標籤裡的文字

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
import time
from selenium.webdriver.common.by import By

s = Service("./chromedriver.exe")
driver = webdriver.Chrome(service=s)

#s = Service("./geckodriver.exe")
#driver = webdriver.Firefox(service=s)

driver.get("https://www.yes123.com.tw/wk_index/joblist.asp?search_job_t=1&strrec=0&search_key_word=%E5%BE%8C%E7%AB%AF%E5%B7%A5%E7%A8%8B%E5%B8%AB&search_type=job&search_item=1&search_from=index")
titles=driver.find_elements(by=By.TAG_NAME, value="h5")
for title in titles:
    print(title.text)
time.sleep(2)
driver.close() 
```

### 模擬滑鼠 移動(失敗)

check 滑鼠移動/點網址/印出浮動視窗

```python
from selenium.webdriver import ActionChains
s = Service("./chromedriver.exe")
driver = webdriver.Chrome(service=s)
driver.get("https://www.yes123.com.tw/wk_index/joblist.asp?search_job_t=1&strrec=0&search_key_word=%E5%BE%8C%E7%AB%AF%E5%B7%A5%E7%A8%8B%E5%B8%AB&search_type=job&search_item=1&search_from=index")
# titles=driver.find_elements_by_class_name("Job_opening_item")#舊寫法
titles=driver.find_element(by=By.XPATH, value="/html/body/div[3]/div[2]/div/div[1]/div[1]/div[9]/div[1]/div[2]/div[1]/h5/a")
ActionChains(driver).move_to_element(titles).perform()
time.sleep(2)
ttitles=driver.find_element(by=By.CLASS_NAME, value="row.job_refer_detail_item")
print(ttitles.text)
time.sleep(2)

driver.close() 
```

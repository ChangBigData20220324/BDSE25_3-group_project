### for loop 印出整頁所有資訊

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
from selenium.webdriver import ActionChains
s = Service("./chromedriver.exe")
driver = webdriver.Chrome(service=s)
driver.get("https://www.yes123.com.tw/wk_index/joblist.asp?search_job_t=1&strrec=0&search_key_word=%E5%BE%8C%E7%AB%AF%E5%B7%A5%E7%A8%8B%E5%B8%AB&search_type=job&search_item=1&search_from=index")
# titles=driver.find_elements_by_class_name("Job_opening_item")#舊寫法



title=driver.find_elements(by=By.CLASS_NAME, value='Job_opening_item')
for titles in title:
    print(titles.text)
    ActionChains(driver).move_to_element(titles).perform()
    time.sleep(1)
    titles=driver.find_element(by=By.CLASS_NAME, value="job_refer_detail.col-4.box_detail_right")
    print(titles.text)
    time.sleep(1)



driver.close() 
```
